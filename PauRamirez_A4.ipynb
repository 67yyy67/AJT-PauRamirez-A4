Exercici 3.

```python
def Fibonacci(n):
    fib = []
    for a in range(n):
        if a == 0:
            fib += [0]
        elif a == 1:
            fib += [1]
        else:
            fib += [fib[a - 1] + fib[a - 2]]
    return fib

print(Fibonacci(10))
```

```python
from IPython.display import clear_output

def mostra_taulell(taulell, diccionari_jugadors, N):
   
    print("    " + "   ".join(str(i) for i in range(N)))
    for fila in range(N):
        contingut = " | ".join(diccionari_jugadors[taulell[fila][columna]] for columna in range(N))
        print(f" {fila}  {contingut}")
        if fila < N - 1:
            print("   " + "---+" *(N - 1) + "---")
    print()
    return


def demana_posicio(taulell, N):
    try:
        fila = int(input(f"Dona'm la fila on vols posar la peça (valor entre 0 i {N - 1})"))
        while fila < 0 or fila > N - 1:
            fila = int(input(f"Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i {N - 1})"))
    except:
        return None, None
    try:
        columna = int(input(f"Dona'm la columna on vols posar la peça (valor entre 0 i {N - 1})"))
        while columna < 0 or columna >  N - 1:
            columna = int(input(f"Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i {N - 1})"))
    except:
        return None, None
    while taulell[fila][columna] != 0:
       fila = int(input(f"Posició ocupada, siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i {N - 1})"))
       columna = int(input(f"Siusplau, dona'm be la columna on vols posar la peça (valor entre 0 i {N - 1})"))
    return fila, columna

def actualitza_taulell(taulell, fila, columna, jugador):
    taulell[fila][columna] = jugador
    return taulell

def canvi_jugador(jugador):
    return jugador*(-1)

def comprovar_victoria(taulell,jugador, N):
    # Victòria per files
    for fila in range(0, N):
        suma = 0
        for columna in range(0,N):
            suma += taulell[fila][columna]
        if suma == N*jugador:
            return True
    # Victòria per columnes
    for columna in range(0, N):
        suma = 0
        for fila in range(0, N):
            suma += taulell[fila][columna]

        if suma == N*jugador:
            return True 
    # Victòria per diagonals
    suma = 0
    for diagonal in range(0,N):
        suma += taulell[diagonal][diagonal]            
    if suma == N*jugador:
        return True 
    # Victòria per diagonal
    suma = 0
    for diagonal in range(0,N):
        suma += taulell[diagonal][(N - 1)-diagonal]            
    if suma == N*jugador:
        return True 
    
    return False
```

```python
jugador = 1
diccionari_jugadors = {
    0: " ",
    1: "X",
    -1: "O"
    }

condicio_victoria = False
comptador = 0
N = int(input("introdueix el nombre de files i columnes del taulell (nombre natural > 1)"))
taulell = [[0 for a in range(N)] for n in range(N)]
while not condicio_victoria and comptador < (N ** 2):
    clear_output(wait=True)
    mostra_taulell(taulell,diccionari_jugadors, N)
    fila, columna = demana_posicio(taulell, N)
    if fila == None and columna == None:
        print("Has de donar un nombre natural entre 0 i 2. Sortim del joc")
        break
    taulell = actualitza_taulell(taulell, fila, columna, jugador)
    condicio_victoria = comprovar_victoria(taulell, jugador, N)
    if condicio_victoria:
        clear_output(wait=True)
        mostra_taulell(taulell,diccionari_jugadors, N)
        print(f"Ha guanyat el jugador {diccionari_jugadors[jugador]}")
        break

    comptador += 1
    jugador = canvi_jugador(jugador)
```
